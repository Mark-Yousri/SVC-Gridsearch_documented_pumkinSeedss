```python
df['Class'] = df['Class'].map({'Çerçevelik':0, 'Ürgüp Sivrisi':1 }).astype(int)
```

replacing strings classes with numerical values

```python
from sklearn.model_selection import train_test_split
X = df.drop(['Class'],axis=1)
y=df['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.35, random_state =  101)
```

splitting data into train &test

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state =83 ,max_iter=3000)
classifier.fit(X_train, y_train)
```

Importing logistic regression

```python
y_pred_2 = classifier.predict(X_test)
```

```python
y_pred_1 = classifier.predict(X_train)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred_2)
print (cm)
```

```python
from sklearn.metrics import classification_report
cr=classification_report(y_test, y_pred_2)
print(cr)
```

```python
from sklearn.metrics import classification_report
cr=classification_report(y_train, y_pred_1)
print(cr)
```

```python
from sklearn.svm import SVC
kernel=['linear','rbf','poly','sigmoid']
# Building a Support Vector Machine on train data
for i in kernel :
    svc_model = SVC(C= .1, kernel=i)
    svc_model.fit(X_train, y_train)
    print('')
    print('for kernel:',i)
    print('')
    print('Accuracy is :')
    print('')
    print(svc_model.score(X_train, y_train))
    print(svc_model.score(X_test, y_test))
```

Impoting SVC


1.   It was discovered after many trial that the linear kernel is best
2.   Poly kernel took too much time 
3.sigmoid & RBF kernels gave very low accuracy even after attenuating C & gamma







- Now we try hyper parameter tuning using grid search for C , Kernel & degree
-Warning for this step better use GPU as it is very time consuming

```python
from sklearn.model_selection import GridSearchCV
param_grid={'C':[0.1,1,100,1000],'kernel':['linear','rbf','poly','sigmoid'],'degree': [1,2,3,4,5,6]}
grid=GridSearchCV(SVC(),param_grid)
grid.fit(X_train,y_train)
```

```python
print(grid.best_params_)
print(grid.score(X_test,y_test))
```

- The grid search had chosen from all different combinations the above hyper parameters

```python
svc_model = SVC(C=100, kernel='poly', degree= 3)
svc_model.fit(X_train, y_train)
print(svc_model.score(X_train, y_train))
print(svc_model.score(X_test, y_test))
```
